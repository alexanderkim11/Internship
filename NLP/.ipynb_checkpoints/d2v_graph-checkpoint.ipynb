```python
from pyvis.network import Network
import networkx as nx
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

# Breadth First Search:

```python
"""
This function is a modified version of Breadth First Search that updates the group number of a node based on which
connected component of the graph it belongs to.  The variable "adj" is the adjacency list for the graph, given in the 
Update Node Groups section below.

Function: BFS(start,group)

Parameters:
    start = Node ID as int, represents the node where the BFS will begin traversing from
    group = Group ID to assign to all nodes that BFS finds
    
Output: None, directly modifies nx_graph

"""

#TODO: Add functionality to handle any supplied graph, not just nx_graph

def BFS(start, group):
    inqueue_or_done = {}
    queue = []
    
    queue.append(start)
    inqueue_or_done[start] = True

    while queue:
        cur = queue.pop(0)
        nx_graph.nodes[cur]['group'] = group
        nodes_copy.pop(cur)
        for node in adj[cur]:
            if inqueue_or_done.get(node) == None:
                queue.append(node)
                inqueue_or_done[node] = True
    
```

# Helper Function:

```python
"""
This function returns the title of a document given its index


Function:gettitle(index)

Parameters:
    index = index of document in documents_df as an int
    
Output: Name of document at index, Type: String
"""

def gettitle(index):
    return documents_df.iloc[index].filename
```

# Recieve Input and Process CSV Files:

<b>WARNING:</b>
The functionality of the graphs depreciates somewhat with a large number of groups.  Setting a small minimum cluster size may result in a larger number of groups being displayed.  Pyvis networks do not have many built-in colors, and a large number of groups results in the same color being applied to multiple components.  This then makes it hard to visually differentiate whether or not two components belong to the same group or not.

```python
threshold = float(input("Enter minimum similarity threshold: "))
min_cluster_size = int(input("Enter minimum cluster size to display: "))
```

```python
"""
This cell reads in both documents_df and document_embeddings, removes the filename column in document_embeddings to maintain
functionality with cosine_similarity, then calculates cosine_similarity and stores the resulting DataFrame in newdf2

"""

#TODO: Add Euclidean Distance Option
documents_df = pd.read_csv("document_df.csv")
tempdf = pd.read_csv("document_embeddings.csv")
tempdf = tempdf.drop(["filename"], axis = 1)
document_embeddings = tempdf.values
pairwise_similarities=cosine_similarity(document_embeddings)
newdf2 = pd.DataFrame(pairwise_similarities)
newdf2
```

# Build Network:

```python
"""
This cell builds out the intial graph using the cosine similarities.  Only nodes that have an edge with a cosine similarity
greater than the specified threshold get added to the graph.  The Networkx graph is then temporarily assigned to a Pyvis
network graph in the last line in order to use the .get_adj_list() method in the next cell

"""


net = Network(height='100%', width='100%') 
nx_graph = nx.Graph()
nodes = {}
for i in range(len(newdf2.index)):
    title = gettitle(i)
    similar_ix=np.argsort(pairwise_similarities[i])[::-1]
    for sim in similar_ix:
        if sim==i:
            continue
        elif pairwise_similarities[i][sim] < threshold:
            break
        else:
            if nodes.get(i) == None:
                nx_graph.add_node(i, title=title, size=20, group = 0)
                nodes[i] = 0
            sim = sim.item()
            sim_title = gettitle(sim)
            if nodes.get(sim) == None:
                nx_graph.add_node(sim, title=sim_title,  size=20, group = 0)
                nodes[sim] = 0
            value = pairwise_similarities[i][sim]
            nx_graph.add_edge(i, sim, title = str(value))
net.from_nx(nx_graph)
```

# Update Node Groups Based on Threshold Values and Connected Component:

```python
"""
This cell gets the adjacency list from the graph above, then uses BFS to update the group number (and consequently the color)
of every node based on which connected component it belongs to.

"""

adj = net.get_adj_list()
nodes_copy = nodes
group = 1
while len(nodes_copy) > 0:
    start = list(nodes_copy.keys())[0]
    BFS(start, group)
    group += 1
```

# Create and Display New Graph Based On Minimum Cluster Size and New Node Groupings:

```python
"""
This cell gets a count of how many nodes are in each group, then assigns it to a DataFrame and displays it
"""


group_counts = {}
for item in nx_graph.nodes.values():
    if group_counts.get(item["group"]) == None:
        group_counts[item["group"]] = 1
    else:
        group_counts[item["group"]] += 1
srt = sorted(group_counts.items(), key = lambda kv: kv[1], reverse = True)
df = pd.DataFrame(srt)
ind = list(df[0])
val = list(df[1])
counts = pd.DataFrame(val, index = ind, columns = ["Node Count"])
counts.index.name = "Group Number"
counts
```

```python
"""
This cell gets a list of all the nodes from nx_graph (the one that's already been modified to include the proper group numbers)
that fall into one of the top groups with populations above the specified minimum cluster size.  It then adds a tuple 
containing the node ID and the node itself to a list called top_nodes.
"""
top = list(counts[counts["Node Count"] >= min_cluster_size].index)
top_nodes = []
top_node_num = []
for node in list(nx_graph.nodes):
    if nx_graph.nodes[node]["group"] in top:
        top_nodes.append((node, nx_graph.nodes[node]))
        top_node_num.append(node)
```

```python
"""
This cell does the exact same thing as the cell above, just with edges instead of nodes.  The list top_edges just contains the
actual edge object, no such thing as edge_id
"""

top_edges = []
for edge in list(nx_graph.edges):
    if edge[0] in top_node_num and edge[1] in top_node_num:
        top_edges.append(edge)
```

```python
"""
This cell builds out a new graph using the top_nodes and top_edges from above, then displays it in an interactive graph.
"""

nx_graph2 = nx.Graph()
nx_graph2.add_nodes_from(top_nodes)
nx_graph2.add_edges_from(top_edges)
net3 = Network(notebook = True, height='1000', width='1000') 
net3.from_nx(nx_graph2)
net3.show('cluster.html')
```

# Leiden Community Detection:

```python
"""
This cell is largely adapted from the Towards Data Science article found in the Sources section below.  This section reassigns
the group numbers of nodes in nx_graph.  Instead of using BFS to just assign distinct groups to connected components, this
section runs the Leiden CD algorithm to find distinct communities, then assign a unique group to each member of every distinct
community.  More information about how the Leiden CD algorithm works can be found in the articles referenced in the Sources
section.
"""

from cdlib import algorithms
coms = algorithms.leiden(nx_graph)
for i in range(len(coms.communities)):
    for node in coms.communities[i]:
        nx_graph.nodes[node]['group'] = i
```

```python
"""
This cell is the same as above.

This cell gets a count of how many nodes are in each group, then assigns it to a DataFrame and displays it.
"""

group_counts = {}
for item in nx_graph.nodes.values():
    if group_counts.get(item["group"]) == None:
        group_counts[item["group"]] = 1
    else:
        group_counts[item["group"]] += 1
srt = sorted(group_counts.items(), key = lambda kv: kv[1], reverse = True)
df = pd.DataFrame(srt)
ind = list(df[0])
val = list(df[1])
counts = pd.DataFrame(val, index = ind, columns = ["Node Count"])
counts.index.name = "Group Number"
counts
```

```python
"""
This cell is the same as above

This cell gets a list of all the nodes from nx_graph (the one that's already been modified to include the proper group numbers)
that fall into one of the top groups with populations above the specified minimum cluster size.  It then adds a tuple 
containing the node ID and the node itself to a list called top_nodes.
"""

top = list(counts[counts["Node Count"] >= min_cluster_size].index)
top_nodes = []
top_node_num = []
for node in list(nx_graph.nodes):
    if nx_graph.nodes[node]["group"] in top:
        top_nodes.append((node, nx_graph.nodes[node]))
        top_node_num.append(node)
top_edges = []
for edge in list(nx_graph.edges):
    if edge[0] in top_node_num and edge[1] in top_node_num:
        top_edges.append(edge)
```

```python
"""
This cell is the same as above.

This cell builds out a new graph using the top_nodes and top_edges from above, then displays it in an interactive graph.
"""

nx_graph3 = nx.Graph()
nx_graph3.add_nodes_from(top_nodes)
nx_graph3.add_nodes_from(top_nodes)
nx_graph3.add_edges_from(top_edges)
net3 = Network(notebook = True, height='1000', width='1000') 
net3.from_nx(nx_graph3)
net3.show('community.html')
```
